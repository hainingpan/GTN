```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pickle
```

## Results

### Born rule

```python
with open('GTN_a1(0.50,0.50,1)_a2(0.00,0.99,21)_b1(1.00,1.00,1)_b2(1.00,1.00,1)_En200_L64_t64_Born.pickle','rb') as f:
    rs64=pickle.load(f)
```

```python
with open('GTN_a1(0.50,0.50,1)_a2(0.00,0.99,21)_b1(1.00,1.00,1)_b2(1.00,1.00,1)_En200_L128_t128_Born.pickle','rb') as f:
    rs128=pickle.load(f)
```

```python
with open('GTN_a1(0.50,0.50,1)_a2(0.00,0.99,21)_b1(1.00,1.00,1)_b2(1.00,1.00,1)_En200_L256_t256_Born.pickle','rb') as f:
    rs256=pickle.load(f)
```

```python
fig,ax=plt.subplots(figsize=(10,6))
ax.errorbar(np.linspace(*rs64['args'].a2[:2],int(rs64['args'].a2[2])),np.squeeze(rs64['MI'].mean(axis=-1))/np.log(2),np.squeeze(rs64['MI'].mean(axis=-1))/np.sqrt(rs64['args'].es)/np.log(2),capsize=3,fmt='.-',color='r',label='L={:d}'.format(rs64['args'].L))
ax.errorbar(np.linspace(*rs128['args'].a2[:2],int(rs128['args'].a2[2])),np.squeeze(rs128['MI'].mean(axis=-1))/np.log(2),np.squeeze(rs128['MI'].mean(axis=-1))/np.sqrt(rs128['args'].es)/np.log(2),capsize=3,fmt='.-',color='b',label='L={:d}'.format(rs128['args'].L))
ax.errorbar(np.linspace(*rs256['args'].a2[:2],int(rs256['args'].a2[2])),np.squeeze(rs256['MI'].mean(axis=-1))/np.log(2),np.squeeze(rs256['MI'].mean(axis=-1))/np.sqrt(rs256['args'].es)/np.log(2),capsize=3,fmt='.-',color='c',label='L={:d}'.format(rs256['args'].L))

ax.set_ylabel('MI [log2]')
ax.set_xlabel('$a_2$')
ax.legend()
```
